## Feature Engineering

### Subtask:
Create new meaningful features from existing ones that could potentially improve model performance, providing explanations for the new features.


**Reasoning**:
I will create three new features ('V4_Amount_Interaction', 'V17_Amount_Interaction', and 'V4_squared') based on the instructions, using existing columns in the DataFrame `df`. Then, I will display the first few rows of the updated DataFrame to verify the new features.



```python
print("--- Creating New Features ---")

# 1. Create a new feature named 'V4_Amount_Interaction'
df['V4_Amount_Interaction'] = df['V4'] * df['Amount']

# 2. Create a new feature named 'V17_Amount_Interaction'
df['V17_Amount_Interaction'] = df['V17'] * df['Amount']

# 3. Create a new feature named 'V4_squared'
df['V4_squared'] = df['V4'] ** 2

print("New features created successfully.")

# Display the first few rows of the DataFrame to verify the creation of the new features
print("\nDataFrame with new features:")
print(df.head())
```

### Explanation of New Features:

1.  **V4_Amount_Interaction:**
    *   **Creation Method:** This feature is generated by multiplying the `V4` principal component with the `Amount` of the transaction. `V4` showed one of the strongest positive correlations with the `Class` (fraudulent transactions) in the correlation analysis. The `Amount` of the transaction is also a critical factor in fraud detection.
    *   **Rationale:** The interaction between `V4` and `Amount` could capture more complex relationships. It hypothesizes that the impact of `V4` on fraud detection might be amplified or altered depending on the transaction `Amount`. For instance, a high `V4` value combined with a large `Amount` might be a stronger indicator of fraud than either feature alone.

2.  **V17_Amount_Interaction:**
    *   **Creation Method:** This feature is created by multiplying the `V17` principal component with the `Amount` of the transaction. `V17` exhibited the strongest negative correlation with the `Class` (fraudulent transactions) in the correlation analysis.
    *   **Rationale:** Similar to `V4_Amount_Interaction`, this feature aims to capture interaction effects. A lower `V17` value, especially in conjunction with a specific `Amount` range, might provide a more nuanced signal for fraudulent activity. This interaction could help the model distinguish between different types of fraudulent patterns where both `V17` and `Amount` play a combined role.

3.  **V4_squared:**
    *   **Creation Method:** This feature is created by squaring the `V4` principal component.
    *   **Rationale:** Squaring a feature can help capture non-linear relationships with the target variable. If `V4` has a quadratic relationship with the likelihood of fraud, or if extreme values (both positive and negative after scaling) are more indicative of fraud, then `V4_squared` could be a useful feature. It effectively emphasizes larger absolute values of `V4`, regardless of their original sign, which can be beneficial if deviations from the mean in either direction are important for fraud detection.

## Prepare Data for Modeling

### Subtask:
Split the processed data into training and testing sets, ensuring a proper separation for model evaluation.


**Reasoning**:
I will prepare the data for modeling by separating features and the target variable, then split them into training and testing sets using `train_test_split` with specified parameters to ensure proper evaluation.



```python
from sklearn.model_selection import train_test_split

print("--- Preparing Data for Modeling ---")

# 1. Separate features (X) and target variable (y)
X = df.drop('Class', axis=1)
y = df['Class']

print("Features (X) and target (y) separated.")

# 3. Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print("Data split into training and testing sets successfully.")

# 4. Print the shapes of the resulting sets to verify the split
print(f"\nShape of X_train: {X_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")

print("\nClass distribution in original dataset:\n", y.value_counts(normalize=True))
print("\nClass distribution in y_train:\n", y_train.value_counts(normalize=True))
print("\nClass distribution in y_test:\n", y_test.value_counts(normalize=True))
```

## Train and Compare ML Models

### Subtask:
Implement and train at least three different machine learning algorithms (e.g., Logistic Regression, Decision Tree, Random Forest, SVM) and compare their baseline performance using a cross-validation strategy.


**Reasoning**:
I need to import the required libraries and define the models and evaluation metrics as per the instructions, which will set up the environment for training and comparing the models.



```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score
import numpy as np

print("--- Initializing Models and Cross-Validation Setup ---")

# Define models to train
models = {
    'Logistic Regression': LogisticRegression(random_state=42, solver='liblinear', penalty='l1'), # Using liblinear and l1 for better performance on smaller datasets/features
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42)
}

# Define evaluation metrics appropriate for imbalanced classification
# 'recall': ability to find all positive samples
# 'f1': harmonic mean of precision and recall
# 'roc_auc': measure of classifier performance across all possible classification thresholds
scoring_metrics = ['recall', 'f1', 'roc_auc']

# Configure Stratified K-Fold for cross-validation
kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

print("Models and metrics defined. StratifiedKFold configured.")
```

**Reasoning**:
Now that the models and cross-validation setup are defined, I will iterate through each model, perform stratified cross-validation using `cross_val_score` for each of the specified metrics, and then print the mean and standard deviation of the scores to compare their baseline performance.



```python
print("--- Training Models and Evaluating Baseline Performance ---")

results = {}

for model_name, model in models.items():
    print(f"\nProcessing {model_name}...")
    model_scores = {}
    for metric in scoring_metrics:
        scores = cross_val_score(model, X_train, y_train, cv=kf, scoring=metric, n_jobs=-1) # Use all available CPU cores
        model_scores[metric] = scores
        print(f"  {metric.replace('_', ' ').capitalize()} (mean): {scores.mean():.4f} (+/- {scores.std():.4f})")
    results[model_name] = model_scores

print("\nBaseline performance evaluation complete.")
```

**Reasoning**:
The previous execution for the Random Forest model was interrupted, as not all evaluation metrics were printed. I need to re-run the training and evaluation loop to ensure all metrics (F1 and ROC AUC) for the Random Forest model are computed and displayed, and the final completion message is shown.



```python
print("--- Training Models and Evaluating Baseline Performance ---")

results = {}

for model_name, model in models.items():
    print(f"\nProcessing {model_name}...")
    model_scores = {}
    for metric in scoring_metrics:
        scores = cross_val_score(model, X_train, y_train, cv=kf, scoring=metric, n_jobs=-1) # Use all available CPU cores
        model_scores[metric] = scores
        print(f"  {metric.replace('_', ' ').capitalize()} (mean): {scores.mean():.4f} (+/- {scores.std():.4f})")
    results[model_name] = model_scores

print("\nBaseline performance evaluation complete.")
```
